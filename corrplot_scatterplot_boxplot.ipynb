```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import statistics
```

```python
#import dataset
df = pd.read_csv("student-mat.csv")
```

```python
#Standart sapma (https://www.geeksforgeeks.org/python-statistics-stdev/)
print("Yaş için standart sapma:  % s " % (statistics.stdev(df["age"])))
print("Yolculuk süresi için standart sapma:  % s " % (statistics.stdev(df["traveltime"])))
print("Çalışma zamanı için standart sapma:  % s " % (statistics.stdev(df["studytime"])))
print("Final notu için standart sapma:  % s " % (statistics.stdev(df["G3"])))
```

```python
#korelasyon
plt.figure(figsize=(15,8))
sb.heatmap(df.corr(), annot = True)
plt.show()
```

```python
#saçılım grafiği (https://www.geeksforgeeks.org/matplotlib-pyplot-scatter-in-python/)
plt.scatter(df["G3"], df["age"], alpha=0.5,edgecolor ="green")
plt.title('Scatter plot ')
plt.xlabel("G3")
plt.ylabel("age")
plt.show()
```

```python
import matplotlib.pyplot as plt
 
plt.scatter(df["G3"], df["age"], c ="green",
            linewidths = 2,
            marker ="s",
            edgecolor ="green",
            s = 50)
 
plt.scatter(df["G3"], df["studytime"], c ="yellow",
            linewidths = 2,
            marker ="^",
            edgecolor ="yellow",
            s = 200)
plt.scatter(df["G3"], df["absences"], c ="red",
            linewidths = 2,
            marker ="o",
            edgecolor ="red",
            s = 10)
plt.scatter(df["G3"], df["G1"], c ="purple",
            linewidths = 2,
            marker ="x",
            edgecolor ="purple",
            s = 50)
 
plt.xlabel("final notu")
plt.ylabel("Y-axis")
plt.show()
```

```python
#kutu grafiği
data=[df["age"],df["studytime"],df["goout"],df["absences"],df["G3"]]
fig,ax=plt.subplots()
ax.set_title('Boxplot grafiği')
ax.boxplot(data,patch_artist = True,notch ='True')
```
