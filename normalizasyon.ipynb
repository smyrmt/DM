```python
import numpy as np
import pandas as pd
import sklearn.metrics as mt
from sklearn import preprocessing
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import dataset
df=pd.read_csv("student-mat.csv")
print(df.shape)
df.head()
```

```python
#normalizasyon iÃ§in verileri kopyalayalÄ±m 1. ve 2. dÃ¶nem notlarÄ± iÃ§in normalizasyon
df_norm=df.copy()
df_norm[['G1','G2']]=preprocessing.minmax_scale(df[['G1','G2']])
#sonuÃ§larÄ± gÃ¶relim
print(df_norm[['G1','G2']].values)
```

```python
df_std=df.copy()
st_olcek=preprocessing.StandardScaler()
df_std[['G1','G2']]=st_olcek.fit_transform(df[['G1','G2']])
print(df_std[['G1','G2']].values)
```

```python
#Histogram 
plt.hist(df.G1,bins=12, color='lightblue',label="1. dÃ¶nem")
plt.xlabel("Notlar")
plt.ylabel("Frekans")
plt.title("1. dÃ¶nem notlarÄ±")
plt.show()
plt.hist(df.G2,bins=12, color='violet',label="2. dÃ¶nem")
plt.xlabel("Notlar")
plt.ylabel("Frekans")
plt.title("2. dÃ¶nem notlarÄ±")
plt.show()
```

```python
db=df.copy()
```

```python
#1. dÃ¶nem notlarÄ± iÃ§in Q1, Q2(median), Q3 IQR
Q1=np.percentile(db[["G1"]],25,interpolation='midpoint')
Q2=np.percentile(db[["G1"]],50,interpolation='midpoint')
Q3=np.percentile(db[["G1"]],75,interpolation='midpoint')
iqr=Q3-Q1
print("1. dÃ¶nem verileri")
print("medyan :",Q2)
print("q1     :",Q1)
print("q3     :",Q3)
print("iqr    :",iqr)
```

```python
#2. dÃ¶nem notlarÄ± iÃ§in Q1, Q2(median), Q3 IQR
Q1=np.percentile(db[["G2"]],25,interpolation='midpoint')
Q2=np.percentile(db[["G2"]],50,interpolation='midpoint')
Q3=np.percentile(db[["G2"]],75,interpolation='midpoint')
iqr=Q3-Q1
print("2. dÃ¶nem verileri")
print("medyan :",Q2)
print("q1     :",Q1)
print("q3     :",Q3)
print("iqr    :",iqr)
```

```python
data=[db["G1"],db["G2"]]
fig,ax=plt.subplots()
ax.set_title('1. dÃ¶nem ve 2. dÃ¶neme ait notlarÄ±n Boxplot grafiÄŸi')
ax.boxplot(data,patch_artist = True,notch ='True')
```
